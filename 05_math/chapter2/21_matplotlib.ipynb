### 2.2.1 Mark Points on Your Graph

```python
import matplotlib.pyplot as plt
x_numbers = [1, 2, 3]
y_numbers = [2, 4, 6]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 10))

axes[0].set_aspect(0.5)
axes[0].plot(x_numbers, y_numbers)
axes[1].set_aspect(1)
axes[1].plot(x_numbers, y_numbers, marker='o') 
axes[2].set_aspect(0.5)
axes[2].plot(x_numbers, y_numbers, 'o')

plt.show()
```
