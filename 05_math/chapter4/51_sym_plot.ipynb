## 4.5 Plotting using symPy

### Case 1

```python
from sympy.plotting import plot
from sympy import Symbol

x = Symbol('x')
expr = (x-10)*(x+5)*(x-2)
plot(expr)
```

### Case 2

```python
from sympy.plotting import plot
from sympy import Symbol

x = Symbol('x')
fx = 2*x + 2
plot(fx, (x, -5, 5), title='A Line', xlabel='x', ylabel='fx = 2x+3')
```

### 4.5.1 Plotting Expression Input by the User

```python
from sympy import sympify, Symbol, solve
from sympy.plotting import plot


def plot_expression(expr):
    y = Symbol('y')
    solution = solve(expr, y)
    expr_y = solution[0]
    plot(expr_y)

if __name__ == '__main__':
    expr = input('Enter your expression in terms of x and y: ')

    try:
        expr = sympify(expr)
    except SympifyError:
        print('Invalid input')
    else:
        plot_expression(expr)
```

### 4.5.2 Plotting Multiple Functions

```python
from sympy.plotting import plot
from sympy import Symbol

x = Symbol('x')
p = plot(2*x+4, 3*x-6, legend=True, show=False)
p[0].line_clor = 'r'
p[1].line_colr = 'b'
p.show()
```
