### 6.1.1 Drawing a Circle

```python
''' 
Example of using matplotlib's Circle patch
'''

import matplotlib.pyplot as plt

def create_circle():
    # circle = plt.Circle((0, 0), radius=0.5)
    rectangle = plt.Rectangle((0, 0), width=1, height=1,angle=45)
    return rectangle

def show_shape(patch):
    ax = plt.gca()
    ax.add_patch(patch)
    plt.axis('scaled')
    plt.show()

if __name__ == '__main__':
    c = create_circle()
    show_shape(c)
```

### Triangle

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Coordinates of our triangle
pts = np.array([[3,3], [2,5], [5,np.sqrt(5**2 - 2**2)]])
p = Polygon(pts)
ax = plt.gca()
ax.add_patch(p)
ax.set_xlim(1,7)
ax.set_ylim(1,8)
plt.show()
```
